```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def vif_calc():
    vif=pd.DataFrame()
    vif["VIF FACTOR"]=[variance_inflation_factor(d.values,i)for i in range(d.shape[1])]
    vif["FEATURES"]= d.columns
    print(vif)
    
vif_calc()
```

# Testing on various models

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y_train,y_test=train_test_split(x1,y,test_size=.33,random_state=42)
from sklearn import metrics
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import math
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor

lr=LinearRegression()
dtr=DecisionTreeRegressor()
Knr=KNeighborsRegressor()
svr=SVR()

rfr=RandomForestRegressor(n_estimators=100,)
ada=AdaBoostRegressor()
gbr=GradientBoostingRegressor()
```

```python
model=[lr,dtr,Knr,svr,rfr,ada,gbr]
for m in model:
    m.fit(x1_train,y_train)
    score=m.score(x1_train,y_train)
    pred=m.predict(x1_test)
    print('model score',m,'is:',score)
    mse=metrics.mean_squared_error(y_test,pred)
    mae=metrics.mean_absolute_error(y_test,pred)
    rmse=math.sqrt(metrics.mean_squared_error(y_test,pred))
    r2=r2_score(y_test,pred)
    print("mean squared error:",mse)
    print("mean absolute error:",mae)
    print("root mean squared error",rmse)
    print("accuracy of model:",m,r2)
    print('\n')
```

# k fold cross validation

```python
from sklearn.model_selection import cross_val_score
model=[lr,dtr,Knr,svr,rfr,ada,gbr]
for m in model:
    score=cross_val_score(m,x1,y,cv=5)
    print(score)
    print("accuracy of model",m,'is:',score.mean())
    print(score.std())
    print("\n")
```

after averaging the accuracy score from train_test_split and cross_val_score the best selectd model is Gradient Boosting regressor

# hypertuning with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
parameters={"loss":['squared_error', 'absolute_error', 'huber', 'quantile'],
           'max_features':["auto",'sqrt','log2'],
           'max_depth': np.arange(2,10)}
gsv=GridSearchCV(gbr,parameters,cv=3)
gsv.fit(x1_train,y_train)
print("for this",gsv.best_params_)
```

# final model

```python
# Initialize the GradientBoostingRegressor with max_features as an integer
gbr = GradientBoostingRegressor(loss='huber', max_depth=8, max_features=5)

# Fit the model to the training data
gbr.fit(x1_train, y_train)

# Make predictions on the test data
predgbr = gbr.predict(x1_test)

# Calculate various evaluation metrics
score = gbr.score(x1_train, y_train)
r2 = r2_score(y_test, predgbr)
mse = metrics.mean_squared_error(y_test, predgbr)
mae = metrics.mean_absolute_error(y_test, predgbr)
rmse = math.sqrt(mse)

print("Model score is:", score)
print('Model accuracy is:', r2)
print("Mean squared error:", mse)
print("Mean absolute error:", mae)
print("Root mean squared error:", rmse)

# Create a DataFrame to display actual and predicted prices
dfr = pd.DataFrame({"Actual price": y_test, "Predicted price": predgbr})
dfr
```
